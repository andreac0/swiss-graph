### Call nodes.py to populate the DB

### Add Articles and references to nodes

```python
from articles import populate_sections_and_citations
import os # Import the os module

# Base directory containing subfolders of JSON files
BASE_FINAL_JSON_DIR = "./LawsDocs/processed/JSONs_classified_edges/"
rs_to_ru_mapping = "./raw_data/RS_RU_mapping.json"

# Iterate through subdirectories in the BASE_FINAL_JSON_DIR
for item_name in os.listdir(BASE_FINAL_JSON_DIR):
    item_path = os.path.join(BASE_FINAL_JSON_DIR, item_name)
    if os.path.isdir(item_path):
        print(f"--- Processing subdirectory: {item_path} ---")
        populate_sections_and_citations(item_path, rs_to_ru_mapping, driver)
        print(f"--- Finished processing subdirectory: {item_path} ---")
    else:
        print(f"Skipping non-directory item: {item_path}")

print("All subdirectories processed.")
```
