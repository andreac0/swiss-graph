```python
import fitz

pdf = fitz.open('./LawsDocs/PDFs_PageCodes/1998/2009_01.09_Ordinanza del Consiglio federale.pdf')
out = ""
for page in pdf:
    blocks = page.get_text("dict")["blocks"]

    for block in blocks:
        if "lines" in block:
            for line in block["lines"]:
                for span in line["spans"]:
                    text = span["text"].strip()
                    font_size = span["size"]
                    font_name = span["font"]
                    is_bold = "Bold" in font_name or "bold" in font_name

                    if not text:
                        continue

                    # Detect Headers (Chapters)
                    if font_size > 14:  # Adjust threshold if needed
                        formatted_text = f"# {text}\n"
                    elif font_size > 12:
                        formatted_text = f"## {text}\n"
                    elif is_bold:
                        formatted_text = f"**{text}**"
                    elif text.startswith("Art.") or text.startswith("Allegato"):
                        formatted_text = f"### {text}\n"
                    elif text[0].isdigit():  # Numbered lists
                        formatted_text = f"- {text}"
                    else:
                        formatted_text = text

        out += formatted_text + "\n"

out += "\n"

with open("output.md", "w") as file:
    file.write(out)
```

```python
import pandas as pd
from collections import defaultdict

# Load the CSV file
df_pdf = pd.read_csv("./raw_data/allActsPDF.csv")

# Extract the last two elements after splitting on '/'
df_pdf["processed_value"] = df_pdf["act.value"].astype(str).str.split('/').str[-2:].str.join('/')

# Find duplicate rows based on this processed value
duplicate_indices_pdf = df_pdf[df_pdf.duplicated("processed_value", keep=False)].index

# Group indices by their processed values
duplicate_groups_pdf = defaultdict(list)
for idx, value in df_pdf.loc[duplicate_indices_pdf, "processed_value"].items():
    duplicate_groups_pdf[value].append(idx)

# Convert to DataFrame for saving
grouped_df_pdf = pd.DataFrame([(key, indices) for key, indices in duplicate_groups_pdf.items()], columns=["Processed_Value", "Indices"])

# Save to CSV
grouped_df_pdf.to_csv("duplicate_indices_pdf.csv", index=False)

print(grouped_df_pdf)
```
