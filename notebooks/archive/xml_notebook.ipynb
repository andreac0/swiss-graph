```python
df_xml = pd.read_csv("./raw_data/acts_xml.csv")
df_xml['publicationDate.value'] = pd.to_datetime(df_xml['publicationDate.value'])
df_xml['publicationDate.value'] = df_xml['publicationDate.value'].dt.strftime('%d.%m.%Y')

# download xmls and save them in folders by year with right name format
import os
import urllib.request
for index, row in df_xml.iterrows():
    year = row['publicationDate.value'][6:]
    if not os.path.exists(f"./LawsDocs/XMLs/{year}"):
        os.makedirs(f"./LawsDocs/XMLs/{year}")
    url = row['fileUrl.value']
    filename = row['act.value'].split('/')[-1] + "_" + row['publicationDate.value'][:5] + "_" + row['typeDoc.value']
    urllib.request.urlretrieve(url, f"./LawsDocs/XMLs/{year}/{filename}")
    break # remove in case of rerun, here only to stop the cycle 
```

```python
import xml.etree.ElementTree as ET
import os
import pandas as pd
from lxml import etree

def extract_href_links(xml_file):
    try:
        # Parse XML with error recovery
        parser = etree.XMLParser(recover=True)  # Allows parsing even if XML is broken
        tree = etree.parse(xml_file, parser)
        root = tree.getroot()

        # Extract all href links
        links = root.xpath("//@href")

        return links

    except etree.XMLSyntaxError as e:
        print(f"XML Syntax Error in file {xml_file}: {e}")
        return []

def get_hrefs_from_folder(folder):
    hrefs = []
    for file in os.listdir(folder):
        hrefs += extract_href_links(os.path.join(folder, file))
    return hrefs

hrefs = []
for year in os.listdir("./LawsDocs/XMLs"):
    year_folder = os.path.join("./LawsDocs/XMLs", year)
    if os.path.isdir(year_folder):
        hrefs += get_hrefs_from_folder(year_folder)

df_hrefs = pd.DataFrame(hrefs, columns=['href'])
df_hrefs.to_csv("./raw_data/hrefs.csv", index=False)
```
